```python
import pandas as PD
from config import sql_id as ID
from sqlalchemy import create_engine
```

```python
port = '5432'
db = 'employees_db'
pgdb = f'postgresql://{ID}@localhost:{port}/{db}'
sql_engine = create_engine(pgdb)
sql_engine.connect();
```

```python
sql_query = 'SELECT * FROM salaries;'
sql_query
```

```python
emp_salary = PD.read_sql_query(sql_query, sql_engine)
emp_salary.head(10)
```

```python
emp_salary.plot(kind='hist', y='salary', title="Histogram of Employee's Salary");
```

```python
sql_query = 'SELECT "Title", AVG("Salary") AS "Average Salary"'
sql_query +=' FROM employee_view'
sql_query +=' GROUP BY "Title"'
sql_query +=' ORDER BY "Title";'
sql_query
```

```python
title_avg = PD.read_sql_query(sql_query, sql_engine)
title_avg
```

```python
title_avg.plot(kind='bar', x='Title', y='Average Salary', title='Average Salary By Title', legend='');
```
